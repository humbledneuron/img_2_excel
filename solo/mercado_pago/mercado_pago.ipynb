```python
#turn gray letters to red but this is a utter flop 
import cv2
import numpy as np

def colorize_specific_gray_text(image_path, output_path, target_color, red_color):
    try:
        # Load the image
        image = cv2.imread(image_path)
        
        if image is None:
            raise FileNotFoundError(f"Error: Couldn't read the image file '{image_path}'")
        
        # Convert the image to grayscale
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        # Define the threshold value for the target color
        threshold_value = 10
        
        # Check each pixel in grayscale image
        for y in range(gray.shape[0]):
            for x in range(gray.shape[1]):
                # Get the pixel value at (x, y) in grayscale image
                pixel_value = gray[y, x]
                
                # Check if the pixel value is close to the target color
                if np.abs(pixel_value - target_color) <= threshold_value:
                    # Apply red color to the corresponding pixel in BGR image
                    image[y, x] = red_color
        
        # Save the output image
        cv2.imwrite(output_path, image)
        
        print(f"Specific gray text in '{image_path}' turned into red and saved as '{output_path}'.")
    
    except Exception as e:
        print(f"Error: {e}")

# Provide the input and output paths
input_image_path = 'sample1.jpg'
output_image_path = 'output_image.jpg'

# Define the target color in grayscale (for #E3E3E3, it's approximately 227)
target_color = 227

# Define the red color in BGR format
red_color = [0, 0, 255]

# Process the image and save the result
colorize_specific_gray_text(input_image_path, output_image_path, target_color, red_color)
```

```python
import cv2
import pytesseract
import pandas as pd
from openpyxl import Workbook

# Load the image using OpenCV
image_path = 'sample1.jpg'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Use pytesseract to extract text from the grayscale image
extracted_text = pytesseract.image_to_string(gray_image)

# Split the extracted text into lines
lines = extracted_text.split('\n')

# Process the lines to extract key-value pairs and amounts
data = []
amounts = []
for line in lines:
    if ':' in line:
        key, value = line.split(':', 1)
        key = key.strip()
        value = value.strip()
        if '$' in value:
            amount = value.split('$')[-1].strip()
            amounts.append(amount)
        else:
            amounts.append('')  # Placeholder for lines without amounts
        data.append([key, value])

# Create a Pandas DataFrame
df = pd.DataFrame(data, columns=['Field', 'Value'])

# Add the amounts to the DataFrame as a new column
df['Amounts'] = amounts

# Create an Excel workbook and sheet
wb = Workbook()
ws = wb.active

# Write headers to Excel
headers = ['Field', 'Value', 'Amounts']
ws.append(headers)

# Write data to Excel
for row in data:
    ws.append(row)

# Save the Excel file
wb.save('extracted_info_with_amounts.xlsx')
```

```python
#takes some files are extracted and save the data
import cv2
import pytesseract
import re
import os
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment

# Function to extract data from an image and return the extracted information
def extract_data_from_image(image_path):
    # Load the image using OpenCV
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Use pytesseract to extract text from the grayscale image
    extracted_text = pytesseract.image_to_string(gray_image)

    # Define regular expression patterns for different types of data to extract
    date_pattern = r'\b\d{1,2} de [a-z]+ \d{4}\b'
    amount_pattern = r'\$\s*\d+(?:[.,]\d+)?'
    payer_name_pattern = r'(?:de )?([A-Z][a-z]+(?: [A-Z][a-z]+)*)'
    cuit_pattern = r'\b\d{2}-\d{8}-\d{1}\b'
    proof_number_pattern = r'\b\d{11}\b'

    # Find all occurrences of dates, amounts, and CUIT numbers in the extracted text
    dates_found = re.findall(date_pattern, extracted_text)
    amounts_found = re.findall(amount_pattern, extracted_text)
    payer_name_found = re.findall(payer_name_pattern , extracted_text)
    cuit_found = re.findall(cuit_pattern, extracted_text)
    proof_number_found = re.findall(proof_number_pattern, extracted_text)

    # Extract the first date, amount, and CUIT number if any are found
    date = dates_found[0] if dates_found else None
    amount = amounts_found[0] if amounts_found else None
    payer = payer_name_found[1] if payer_name_found else None
    cuit = cuit_found[0] if cuit_found else None
    proof_number = proof_number_found[0] if proof_number_found else None

    # Return the extracted information as a dictionary
    return {
        'BANCO': 'mercado pago',
        'FECHA': date,
        'IMPORTE': amount,
        'TITULAR': payer,
        'CUIT': cuit,
        'NRO COMPROBANTE': proof_number
    }

# Create a new Excel workbook
wb = Workbook()
ws = wb.active
ws.title = "Extracted Data"

# Add headers to the Excel sheet
headers = [ 'FECHA', 'IMPORTE', 'NRO COMPROBANTE', 'TITULAR', 'CUIT', 'BANCO']
ws.append(headers)


last_row = ws.max_row

for row in range(2, last_row + 1):  # Assuming data starts from row 2
    importe_formula = f'=SUM(VALUE(TRIM(SUBSTITUTE(B{row},"$",""))))'
    ws[f'B{row}'] = importe_formula  # Assuming the result should be in column C


# Calculate the total sum for the 'Importe' column
total_importe_formula = f'=SUM(C2:C{last_row})'
ws[f'C{last_row + 1}'] = total_importe_formula  # Assuming the total sum goes in the row after the last data row


# Apply bold style to the header cells
bold_font = Font(bold=True)
for cell in ws[1]:
    cell.font = bold_font

# Apply padding to all cells
for row in ws.iter_rows():
    for cell in row:
        cell.alignment = Alignment(horizontal='center', vertical='center', wrap_text=True)

# Define the folder containing the images
folder_path = 'mercado_pago_assets/multi'

# Process each image in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
        # Extract data from the image
        image_path = os.path.join(folder_path, filename)
        extracted_data = extract_data_from_image(image_path)

        # Add the extracted data to the Excel sheet
        row_data = [extracted_data[header] for header in headers]
        ws.append(row_data)

# Automatically adjust column widths based on content
for column_cells in ws.columns:
    max_length = 0
    for cell in column_cells:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(cell.value)
        except:
            pass
    adjusted_width = (max_length + 2) * 1.2  # Adjust the multiplier as needed
    ws.column_dimensions[column_cells[0].column_letter].width = adjusted_width

# Save the Excel file
wb.save('mercado_pago_extracted_info.xlsx')
```

```python
import cv2
import pytesseract
import re
import os
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment
from openpyxl.utils import get_column_letter

# Function to extract data from an image and return the extracted information
def extract_data_from_image(image_path):
    # Load the image using OpenCV
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Use pytesseract to extract text from the grayscale image
    extracted_text = pytesseract.image_to_string(gray_image)

    # Define regular expression patterns for different types of data to extract
    date_pattern = r'\b\d{1,2} de [a-z]+ \d{4}\b'
    amount_pattern = r'[\$¢]\s*\d[\d,\.]+' #r'[\$¢]\s*\d+\.?\d+'  #r'\$\s*\d[\d,\.]*' #r'\$\s*\d+(?:[.,]\d+)?'
    payer_name_pattern = r'(?:de )?([A-Z][a-z]+(?: [A-Z][a-z]+)*)'
    cuit_pattern = r'\b\d{2}-\d{8}-\d{1}\b'
    proof_number_pattern = r'\b\d{11}\b'

    # Find all occurrences of dates, amounts, and CUIT numbers in the extracted text
    dates_found = re.findall(date_pattern, extracted_text)
    amounts_found = re.findall(amount_pattern, extracted_text)
    payer_name_found = re.findall(payer_name_pattern , extracted_text)
    cuit_found = re.findall(cuit_pattern, extracted_text)
    proof_number_found = re.findall(proof_number_pattern, extracted_text)

    # Extract the first date, amount, and CUIT number if any are found
    date = dates_found[0] if dates_found else None
    amount = amounts_found[0] if amounts_found else None
    amount = amount.replace('¢', '$')
    payer = payer_name_found[1] if payer_name_found else None
    cuit = cuit_found[0] if cuit_found else None
    proof_number = proof_number_found[0] if proof_number_found else None

    # Return the extracted information as a dictionary
    return {
        'BANCO': 'mercado pago',
        'FECHA': date,
        'IMPORTE': amount,
        'TITULAR': payer,
        'CUIT': cuit,
        'NRO COMPROBANTE': proof_number
    }

# Create a new Excel workbook
wb = Workbook()
ws = wb.active
ws.title = "Extracted Data"

# Add headers to the Excel sheet
headers = [ 'FECHA', 'IMPORTE', 'NRO COMPROBANTE', 'TITULAR', 'CUIT', 'BANCO']
ws.append(headers)

# Define the folder containing the images
folder_path = 'mercado_pago_assets'

# Process each image in the folder
total_sum_formula = 0
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
        # Extract data from the image
        image_path = os.path.join(folder_path, filename)
        extracted_data = extract_data_from_image(image_path)

        # Add the extracted data to the Excel sheet
        row_data = [extracted_data[header] for header in headers]
        ws.append(row_data)
        
        from decimal import Decimal
# Update total sum formula dynamically
        if extracted_data['IMPORTE']:
            amt = re.sub(r'[^\$\s0-9.]', '', extracted_data['IMPORTE'])
            num_amt = Decimal(re.sub(r'[^\d.,]', '', amt)) #r'[^\d.]*' #r'\.(\d+\.)+'
            # amt = re.sub(r'\.(\d+\.)+', '', amt)
            num_amt = Decimal(amt)
            total_sum_formula += num_amt
            
            # frmt_amt = f'{num_amt:.8f}'
            # total_sum_formula += float(frmt_amt)

# Append the total sum row to the Excel sheet
total_sum_row = ['TOTAL', f'${total_sum_formula}', None, None, None, None]
ws.append(total_sum_row)

# Save the Excel file
wb.save('mercado_pago_extracted_info.xlsx')
```
